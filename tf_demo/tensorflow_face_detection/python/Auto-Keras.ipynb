```python
import tensorflow as tf

import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D
from keras.layers import Dense, Activation, Dropout, Flatten

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="1"
```

```python
num_classes = 7 #angry, disgust, fear, happy, sad, surprise, neutral
batch_size = 256
epochs = 5
```

```python
with open("fer2013.csv") as f:
    content = f.readlines()
```

```python
lines = np.array(content)
```

```python
num_of_instances = lines.size
print("number of instances: ",num_of_instances)
print("instance length: ",len(lines[1].split(",")[1].split(" ")))
```

```python
x_train, y_train, x_test, y_test = [], [], [], []
```

```python
for i in range(1,num_of_instances):
    try:
        emotion, img, usage = lines[i].split(",")
          
        val = img.split(" ")
            
        pixels = np.array(val, 'float32')
        
        #emotion = keras.utils.to_categorical(emotion, num_classes)
        
        #print(emotion)
    
        if 'Training' in usage:
            y_train.append(emotion)
            x_train.append(pixels)
        elif 'PublicTest' in usage:
            y_test.append(emotion)
            x_test.append(pixels)
    except Exception as ex:
        print(ex)
```

```python
x_train = np.array(x_train, 'float32')
y_train = np.array(y_train, 'float32')
x_test = np.array(x_test, 'float32')
y_test = np.array(y_test, 'float32')

x_train /= 255 #normalize inputs between [0, 1]
x_test /= 255

x_train = x_train.reshape(x_train.shape[0], 48, 48, 1)
x_train = x_train.astype('float32')
x_test = x_test.reshape(x_test.shape[0], 48, 48, 1)
x_test = x_test.astype('float32')

print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
x_train[0].shape
```

```python
import autokeras as ak
```

```python
model = ak.ImageClassifier(path="automodels", verbose=True)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model.fit(x_train, y_train, time_limit=60*60*24)
```

```python
model.final_fit(x_train, y_train, x_test, y_test, retrain=True)
```

```python
score = model.evaluate(x_test, y_test)
```

```python
score
```

```python
predictions = model.predict(x_test)
```

```python
instances = 0
correct_ones = 0

for i in range(0, len(predictions)):
    if i < 5:
        print("Prediction: ",predictions[i],", Actual: ",y_test[i])
    
    if predictions[i] == y_test[i]:
        correct_ones = correct_ones + 1
    
    instances = instances + 1
```

```python
print("correctly classified ones after training: ",correct_ones)
print("total number of instances in the test set: ",instances)
```
